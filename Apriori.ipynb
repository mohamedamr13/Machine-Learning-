```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

    ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

    ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

    ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

   ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

Preparing Data using one hot encoding 


```python
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
df
```

Support 

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules 

frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

Confidence 

```python
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.2) # associate itemsets with confidence over 20%.
```

```python
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```

## Suggested Strategies 


- Have packages containing scarfs and pants 
- Put additional racks of scarfs next to pants and vice versa since they are both frequently purchased together
- Put additional racks of skirts next to pants since it is likely for one who bought pants to buy skirts 

### Applying Apriori on large database 

```python
import csv
datafile = open('Market_Basket_Optimisation.csv', 'r')
datareader = csv.reader(datafile, delimiter=',')
data = []
for row in datareader:
     data.append(row)  
```

```python
te=TransactionEncoder()
te_ary=te.fit(data).transform(data)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
df.isnull().sum().sum()
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
frequent_itemsets=apriori(df, min_support=0.03, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets[ frequent_itemsets['support'] > 0.1 ]
```

```python
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.0000000000000000000001) 
```

```python
analysis = association_rules(frequent_itemsets,metric="lift",min_threshold=1.2) 
analysis[ (analysis['support'] > 0.03) & (analysis['confidence'] > 0.2) & (analysis['lift'] > 1.5)]
```
